##15. How do you add a regression line to a scatter plot in Seaborn for variables x and y in a DataFrame df?

```python
import seaborn as sns
import pandas as pd

# Assuming you have data for 'x' and 'y' in lists or arrays:
x_data = [1, 2, 3, 4, 5]  # Replace with your actual x data
y_data = [2, 4, 1, 3, 5]  # Replace with your actual y data

# Create a pandas DataFrame:
df = pd.DataFrame({'x': x_data, 'y': y_data})

# Now you can create the regplot:
sns.regplot(data=df, x="x", y="y")
```
